```python
from helpers import clean_df

train_df = clean_df('raw-data/train_locs.csv', 'raw-data/train_pbp.csv')
test_df = clean_df('raw-data/test_locs.csv','raw-data/test_pbp.csv')
```

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
from xgboost import XGBClassifier

train_data = train_df.drop(['id', 'is_oreb'], axis=1)
train_target = train_df['is_oreb']

test_data = test_df.drop(['id', 'is_oreb'], axis=1)
test_target = test_df['is_oreb']

pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler()),
    ('regressor', XGBClassifier())
])
pipeline.fit(train_data, train_target)
predictions = pipeline.predict(test_data)
r2 = r2_score(test_target, predictions)
print("R-squared:", r2)
```
